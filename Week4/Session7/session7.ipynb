```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def expFunc(a):
    b = np.empty(100)
    for i in a:
        np.append(b, i * i)
    return b
```

```python
y = expFunc(x)
""" for i in x:
    print(i)
for j in y:
    print(j) """
plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")

plt.show()
```
