```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sinFunc(a):
    return np.sin(a)
```

```python
def cosFunc(a):
    return np.cos(a)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_title("sin(x)")
ax1.plot(x, sinFunc(x))

ax2.set_title("cos(x)")
ax2.plot(x, cosFunc(x))

plt.tight_layout()
plt.savefig("sin_cos_plot.pdf")
plt.show()
```
